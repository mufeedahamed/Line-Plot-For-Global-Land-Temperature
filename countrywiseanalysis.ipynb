```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the preprocessed dataset
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/9/GlobalLandTemperaturesByCountryNew.csv')

# Select the desired country and time period
country = 'Canada' #Cuba, Denmark, Egypt, china, africa, asia, australia, canada
start_date = '2010-01-01'
end_date = '2013-12-31'
selected_data = df[(df['Country'] == country) & (df['dt'] >= start_date) & (df['dt'] <= end_date)]

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(selected_data['dt'], selected_data['AverageTemperature'], color='Olive')
plt.xlabel('Date')
plt.ylabel('Average Temperature (Celsius)')
plt.title(f'Monthly Average Temperature for {country} (2010-2013) in Celsius')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```
